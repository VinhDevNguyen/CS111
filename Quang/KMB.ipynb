# Tạo ra 1 hash list chứa các từ khóa và dữ liệu rỗng (sau khi thành công ta mới đưa các giá trị tương ứng với các khóa)
# chúng ta phải encode các giá trị trọng file thành dạng UTF-8 vì bảng băm trong python chỉ băm các phần tử ở dạng bytes hoặc mã hex

```python
file_name = 'syllables.txt'
hash_list = []

with open(file_name, mode = 'r', encoding = 'utf-8') as file:
    FileInput_goc = file.readlines()
    for i in FileInput_goc:
        i = hashlib.sha256(i.encode()).hexdigest()[:3:]
        hash_list.append(i)
```

```python
hash_list
```

# khi đưa về dạng này sẽ có ít nhiều phần tử trùng nên vì vậy ta sẽ xóa bớt chúng đi

```python
hash_list = list(OrderedDict.fromkeys(hash_list))
```

```python
hash_list
```

# Bây giờ chúng ta sẽ tạo ra 1 thư viện rỗng dựa trên bảng băm đã được chia sẵn ở hash_list

```python
dictionary = {}
dictionary = {key: [] for key in hash_list}
```

```python
dictionary
```

# Sau khi tạo ra được 1 thư viện rỗng ta sẽ tiếp tục dùng bảng băm để chia các giá trị trong file gốc đưa về đúng với các key trong thư viện rỗng chúng ta mới tạo

```python
for tu in FileInput_goc:
    chia_tu = hashlib.sha256(tu.encode()).hexdigest()[:3:]
    dictionary[chia_tu].append(tu)
```

# sau khi làm xong 1 dictionary dựa trên file syllables chúng ta sẽ tìm từng từ bên file bao2 so với syllables vì file bao2 
# tuy nhiên để nhanh hơn ta có thể dùng phép giao 2 dictionary với nhau sẽ nhanh hơn rất nhiều 


```python
inputFileName = 'bao2.txt'
inputFile = open(inputFileName,encoding = 'utf-8')
readInputFile = inputFile.readlines()
hash_list_for_bao2 = []
for idx,word in enumerate(readInputFile): 
    encodeWord = hashlib.sha256(word.upper().encode()).hexdigest()[:3:]
    hash_list_for_bao2.append(encodeWord)
```

```python
hash_list_for_bao2
```

```python
dict_for_bao2 = {key: [] for key in hash_list_for_bao2}
```

```python
keys_dictionary = set(dictionary.keys())
keys_bao2 = set(dict_for_bao2.keys())
```

# Sử dụng phép giao 2 khóa của dictionary gán kết quả tìm được vào 1 list hoặc arry rồi dùng len() để xem kết quả
# import thử viện time để kiểm tra thời gian thực thi

```python
import time

intersection = list(keys_bao2 & keys_dictionary)
print(len(intersection))
start_time = time.time()
print("%sseconds" % (time.time() - start_time))
```
